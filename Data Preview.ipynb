```python
from sas7bdat import SAS7BDAT as sas
import pandas

pandas.set_option('display.max_columns', 999)
pandas.set_option('display.max_rows', 20)
```

```python
# adm3.sas7bdat is 50mb. loading takes a few seconds.
# instead we preview the header and the first ten items
import itertools
with sas('sas_data/adm3.sas7bdat') as f:
    gen = f.readlines()
    headers = next(gen)
    rows = list(itertools.islice(gen, 0, 10))
    dataframe = pandas.DataFrame(rows, columns=headers)
dataframe    
```

```python
with sas('sas_data/cancer3.sas7bdat') as f:
    dataframe = f.to_data_frame()
dataframe
```

```python
# opd3.sas7bdat is 5gb, too damn big to load all into a dataframe
# instead we preview the header and the first ten items
import itertools
with sas('sas_data/opd3.sas7bdat') as f:
    gen = f.readlines()
    headers = next(gen)
    rows = list(itertools.islice(gen, 0, 10))
    dataframe = pandas.DataFrame(rows, columns=headers)
dataframe    
```
